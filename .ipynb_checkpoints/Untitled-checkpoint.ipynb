```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.ensemble import IsolationForest
```

```python
# 1. Read the CSV into a DataFrame (adjust filename/path as needed)
df = pd.read_csv('challenge_txs.csv', parse_dates=['date'])

print(df.shape)

# 4. Drop exact duplicate rows
df.drop_duplicates(inplace=True)

print(df.shape)

# 5. Check for missing values
print("Missing values by column:")
print(df.isnull().sum())


# 1️⃣ Define a helper to get the mode
def fill_mode(s):
    m = s.mode()
    return m.iloc[0] if not m.empty else np.nan

# 2️⃣ First‐level: (holding_party_id, holding_party_category, tx_type)
grp1 = df.groupby(
    ['holding_party_id', 'holding_party_category', 'tx_type']
)['counter_party_country'].transform(fill_mode)
df['counter_party_country'] = df['counter_party_country'].fillna(grp1)

# 3️⃣ Second‐level: (holding_party_id, tx_type)
grp2 = df.groupby(
    ['holding_party_id', 'tx_type']
)['counter_party_country'].transform(fill_mode)
df['counter_party_country'] = df['counter_party_country'].fillna(grp2)

# 4️⃣ Third‐level: (holding_party_id)
grp3 = df.groupby('holding_party_id')['counter_party_country'] \
         .transform(fill_mode)
df['counter_party_country'] = df['counter_party_country'].fillna(grp3)

# 5️⃣ Final fallback: overall mode
global_mode = df['counter_party_country'].mode().iloc[0]
df['counter_party_country'] = df['counter_party_country'].fillna(global_mode)

# 6️⃣ Verify
print("Remaining nulls in counter_party_country:", 
      df['counter_party_country'].isnull().sum())

print("Missing values by column:")
print(df.isnull().sum())

# 9. Show a preview of the cleaned DataFrame
df.head()
```

```python
# 7. Ensure 'amount' is numeric
df['amount'] = pd.to_numeric(df['amount'], errors='coerce')

# 8. Inspect distribution of transaction types
print("\nTransaction type counts:")
print(df['tx_type'].value_counts())

# 8. Inspect distribution of transaction types
print("\holding_party_id counts:")
print(df['holding_party_id'].value_counts())

# Get the number of unique values in 'col1'
num_unique = df['holding_party_id'].nunique()

# Print the result
print(num_unique)

```

```python
import matplotlib.pyplot as plt

# 1️⃣ Identify the top 10 businesses by total amount
top10_ids = (
    df
    .groupby('holding_party_id')['amount']
    .sum()
    .nlargest(1000)
    .index
    .tolist()
)

# 2️⃣ Filter to those businesses and aggregate daily sums
df_top10 = df[df['holding_party_id'].isin(top10_ids)].copy()
df_top10['date_only'] = df_top10['date'].dt.date

daily = (
    df_top10
    .groupby(['date_only', 'holding_party_id'])['amount']
    .sum()
    .unstack('holding_party_id')
    .fillna(0)
)

# 3️⃣ Plot
plt.figure(figsize=(12, 6))
for biz in daily.columns:
    plt.plot(daily.index, daily[biz], label=str(biz))

plt.xlabel('Date')
plt.ylabel('Daily Transaction Amount')
plt.title('Daily Transaction Amount Time Series for Top 10 Businesses')
# plt.legend(title='Business ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#•	Identify businesses which behavior can be considered abnormal.
```

```python
agg = df.groupby('holding_party_id').agg(
    txn_count     = ('a_idx','count'),
    txn_sum       = ('amount','sum'),
    txn_mean      = ('amount','mean'),
    txn_std       = ('amount','std'),
    first_date    = ('date','min'),
    last_date     = ('date','max'),
)
# compute active span in days and txn_rate
agg['active_days'] = (agg['last_date'] - agg['first_date']).dt.days + 1
agg['txn_rate_per_day'] = agg['txn_count'] / agg['active_days']

# fill NaN std=0
agg['txn_std'].fillna(0, inplace=True)

# 3. Simple IQR‐based outlier detection on total volume and rate
def iqr_outliers(series, k=1.5):
    q1, q3 = series.quantile([0.25,0.75])
    iqr = q3 - q1
    lower, upper = q1 - k*iqr, q3 + k*iqr
    return (series < lower) | (series > upper)

agg['outlier_sum_iqr']  = iqr_outliers(agg['txn_sum'])
agg['outlier_rate_iqr'] = iqr_outliers(agg['txn_rate_per_day'])

# 4. Isolation Forest for multivariate detection
features = ['txn_count','txn_sum','txn_mean','txn_std','txn_rate_per_day']
iso = IsolationForest(contamination=0.01, random_state=42)
agg['iforest_score'] = iso.fit_predict(agg[features])
# in sklearn: -1 => anomaly, +1 => normal
agg['outlier_iforest'] = agg['iforest_score'] == -1

# 5. Combine flags
agg['flagged'] = agg[['outlier_sum_iqr','outlier_rate_iqr','outlier_iforest']].any(axis=1)

# 6. View flagged businesses
anomalous = agg[agg['flagged']].sort_values(by='txn_sum', ascending=False)
print(f"Found {len(anomalous)} flagged businesses:")
anomalous[['txn_count','txn_sum','txn_rate_per_day','outlier_sum_iqr','outlier_rate_iqr','outlier_iforest']].shape
```

```python
import numpy as np

# 1. Compute per‐business mean/std and z‑score
df['biz_mean'] = df.groupby('holding_party_id')['amount'].transform('mean')
df['biz_std']  = df.groupby('holding_party_id')['amount'].transform('std')

df['z_score'] = (df['amount'] - df['biz_mean']) / df['biz_std']

# 2. Mark anomalies
threshold = 3
df['anomaly_z'] = df['z_score'].abs() > threshold

# 3. Extract all flagged transactions
anomalies_z = df[df['anomaly_z']].sort_values('z_score', key=lambda s: s.abs(), ascending=False)

print(f"Found {len(anomalies_z)} z‑score anomalies")
anomalies_z.head()
```

```python
import pandas as pd

# Assuming `df` is your cleaned DataFrame with z-score anomalies already computed:
# Columns required: a_idx, date (as datetime), amount, holding_party_id, z_score, anomaly_z

# 1️⃣ Extract the z‑score anomalies
anomalies_z = df[df['anomaly_z']].copy()

# 2️⃣ Summarize anomaly rates per business
biz_summary = (
    df
    .groupby('holding_party_id')
    .agg(
        total_tx=('a_idx','count'),
        anomalies=('anomaly_z','sum')
    )
    .assign(anomaly_rate=lambda x: x['anomalies'] / x['total_tx'] * 100)
)

# 3️⃣ Function to explain a single transaction anomaly
def explain_transaction_z(df, tx_idx):
    row = df[df['a_idx'] == tx_idx].iloc[0]
    biz = row['holding_party_id']
    amt = row['amount']
    z = row['z_score']
    # Percentile of this amount within the business
    amts = df[df['holding_party_id'] == biz]['amount']
    pct = amts.rank(pct=True).loc[row.name]
    hour = row['date'].hour
    freq_hour = (
        df[(df['holding_party_id']==biz) & (df['date'].dt.hour == hour)]
        .shape[0]
        / df[df['holding_party_id']==biz].shape[0]
    )
    return (
        f"Transaction {tx_idx} (business {biz}):\n"
        f"- Amount €{amt:.2f} ({z:.2f}σ from mean, {pct*100:.1f}th percentile)\n"
        f"- Occurred at {hour}:00 ({freq_hour*100:.1f}% of this business’s transactions at that hour)"
    )

# 4️⃣ Function to explain a business’s anomaly profile
def explain_business(df, biz_id):
    stats = biz_summary.loc[biz_id]
    mean_amt = df[df['holding_party_id']==biz_id]['amount'].mean()
    std_amt  = df[df['holding_party_id']==biz_id]['amount'].std()
    return (
        f"Business {biz_id}:\n"
        f"- Total transactions: {stats['total_tx']}\n"
        f"- Anomalies: {int(stats['anomalies'])} "
        f"({stats['anomaly_rate']:.2f}% of all txns)\n"
        f"- Avg amount: €{mean_amt:.2f}, Std dev: €{std_amt:.2f}"
    )

# 5️⃣ Examples
sample_tx = anomalies_z['a_idx'].iloc[0]
print(explain_transaction_z(df, sample_tx))

top_biz = biz_summary.sort_values('anomaly_rate', ascending=False).index[0]
print("\n" + explain_business(df, top_biz))
```

```python
from sklearn.ensemble import IsolationForest

# 1. Engineer a couple of extra features
df['hour']        = df['date'].dt.hour
df['day_of_week'] = df['date'].dt.dayofweek

# 2. Function that fits & predicts for one business
def detect_iso(biz_df, contam=0.01):
    features = biz_df[['amount', 'hour', 'day_of_week']].values
    iso = IsolationForest(
        n_estimators=100,
        contamination=contam,
        random_state=42
    )
    biz_df['anomaly_iforest'] = iso.fit_predict(features)
    return biz_df

# 3. Apply it business‐by‐business
results = []
for biz_id, group in df.groupby('holding_party_id'):
    flagged = detect_iso(group)
    results.append(flagged)

df_iforest = pd.concat(results, ignore_index=True)

# 4. Extract the outliers (label = -1)
anomalies_if = df_iforest[df_iforest['anomaly_iforest'] == -1]
print(f"Found {len(anomalies_if)} isolation‑forest anomalies")
anomalies_if.head()
```
